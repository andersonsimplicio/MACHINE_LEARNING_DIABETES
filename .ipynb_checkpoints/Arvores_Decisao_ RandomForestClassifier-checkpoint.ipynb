```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold,cross_val_score
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import tree
import pandas as pd
import numpy as np
import pickle
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from yellowbrick.classifier import ConfusionMatrix
from yellowbrick.model_selection import FeatureImportances
import matplotlib.pyplot as plt
from pydotplus import graph_from_dot_data
from dtreeviz.trees import dtreeviz
```

```python
with open('database/diabetes_no_stander.pkl','rb') as f:
    x_treinamento,y_treinamento,x_teste,y_teste = pickle.load(f)
```

```python
x_treinamento.shape,y_treinamento.shape
```

```python
x_teste.shape,y_teste.shape
```

## Parâmetro Base

```python
modelo_Dummy = DummyClassifier(random_state=0,strategy="most_frequent")
modelo_Dummy.fit(x_treinamento,y_treinamento)
previsao = modelo_Dummy.predict(x_teste) 
accuracy_score(y_teste, previsao)
```

## Buscando o hiperparâmetros da Arvore

```python
#'max_depth':[5,6,7,8,9],
#'criterion':['entropy','gini'],
#'max_features':[5,6,7,8,9],
#'min_samples_leaf':[5,6,7,8,9,10,11]


params= {'max_depth':[7],
         'criterion':['entropy'],
         'max_features':[8],
         'min_samples_leaf':[11]
        }
kf = KFold(n_splits=10,shuffle=True)
arvore_test = DecisionTreeClassifier(random_state = 0)
cv = GridSearchCV(arvore_test,param_grid=params,scoring='roc_auc',cv=kf,n_jobs=None,return_train_score=True)
cv.fit(x_treinamento,y_treinamento)
```

```python
cv_results = pd.DataFrame(cv.cv_results_)
cv_results.head()
arvore_melhor = cv.best_estimator_
score = cross_val_score(arvore_melhor,x_treinamento,y_treinamento,cv = kf)
media =score.mean()*100
desvio_padrao = score.std()*100
print(f'Accuracy  media {media:.2f}')
print(f'Intervalo Médio {media-2*desvio_padrao:.2f} - {media+2*desvio_padrao:.2f}')
arvore_melhor
```

```python
ax = plt.axes()
ax.errorbar(cv_results['param_max_depth'],cv_results['mean_train_score'],yerr=cv_results['std_train_score'],
           label='Mean $\pm$ 1 SD Traning score')
ax.errorbar(cv_results['param_max_depth'],cv_results['mean_test_score'],yerr=cv_results['std_test_score'],
           label='Mean $\pm$ 1 SD test score')
         
ax.legend()
plt.xlabel('max_depth')
plt.ylabel('ROC AUC')
```

## Previsão da melhor Arvore de Decisão

```python
rotulos_previsores = ['Gestações', 'Glicose Plasmática', 'Pressão Diastolica',
           'Dobra Triceps', 'Insulina sérica', 'IMC', 'Diabetes Familiar','Idade','colesterol','hba1c']

arvore_final = DecisionTreeClassifier(criterion='entropy', max_depth=8, max_features=6,
                       min_samples_leaf=10, random_state=0)
arvore_final.fit(x_treinamento,y_treinamento)
previsao = arvore_final.predict(x_teste)
print(classification_report(y_teste,previsao))
cm = ConfusionMatrix(arvore_final,classes=['Não diabético - 0', 'Diabético - 1'])
cm.fit(x_treinamento,y_treinamento)
cm.score(x_teste,y_teste);
```

```python
viz = FeatureImportances(arvore_final,labels=rotulos_previsores);
viz.fit(x_treinamento, y_treinamento);
viz.show();
```

## Random forest  buscando o hiperparâmetros

```python
#criterion='entropy', max_depth=8, max_features=6,min_samples_leaf=10
#n_estimators':     Teste 1 [26,36,46,56,66,86,96,106] => 106
# 'max_depth':      Teste 2 [8,7,8,9,10,11,12,13,14,15] => 14
#'max_features':    Teste 3 [3,4,,5,6,7,8,9] => 3
#'min_samples_leaf':Teste 4 [9,10,11,12,13] =>9
random_forest_test = RandomForestClassifier(random_state=0)
params= {
         'n_estimators':[106],   
         'max_depth': [14],
         'max_features':[3],
         'min_samples_leaf': [9]
        }
cv = GridSearchCV(random_forest_test,param_grid=params,scoring='roc_auc',cv=kf,n_jobs=None,
                  return_train_score=True)
cv.fit(x_treinamento,y_treinamento)
cv_results = pd.DataFrame(cv.cv_results_)
```

```python
cv_results.head()
random_melhor = cv.best_estimator_
score = cross_val_score(random_melhor,x_treinamento,y_treinamento,cv = kf)
media =score.mean()*100
desvio_padrao = score.std()*100
print(f'Accuracy  media {media:.2f}')
print(f'Intervalo Médio {media-2*desvio_padrao:.2f} - {media+2*desvio_padrao:.2f}')
random_melhor
```

```python
random_forest = RandomForestClassifier(n_estimators=106,criterion='entropy',max_depth=14,max_features=3,
                                       min_samples_leaf=9,random_state=0)
random_forest.fit(x_treinamento,y_treinamento)
previsao = random_forest.predict(x_teste)
cm = ConfusionMatrix(random_forest,classes=['Não diabético - 0', 'Diabético - 1'])
cm.fit(x_treinamento,y_treinamento)
cm.score(x_teste,y_teste)
print(classification_report(y_teste,previsao))
```

```python
viz = FeatureImportances(random_forest,labels=rotulos_previsores);
viz.fit(x_treinamento, y_treinamento);
viz.show();
```
